```python
from matplotlib import rcParams
rcParams['figure.figsize']=15,6
plt.plot(data.Date,data.Close);  # take the value of column Date and 'Close'; Syntax matplotlib.plot(x,y) 
plt.plot(data.Date,data.Open);
plt.title("Share Price variation with respect to time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.ylim(100,1400)
plt.grid(True,color='k',linestyle=':')
plt.legend(["Close","Open"],fontsize="x-large",loc=1);
```
